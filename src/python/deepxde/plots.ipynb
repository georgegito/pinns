```python
import numpy as np
from matplotlib import pyplot as plt
from naca4digit_airfoil import Naca4DigitAirfoil
import pandas as pd
import matplotlib.patches as patches
```

```python
airfoil = Naca4DigitAirfoil(c=1, M=2, P=4, T=12, a=0, offset_x=0, offset_y=0)
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

ax.set_title('NACA 2412')
airfoil.plot(ax)
ax.set_xlim(-.25, 1.25)
ax.set_ylim(-.25, .25)
ax.grid(False)
ax.set_aspect('equal')
plt.savefig('naca_2412.png', dpi=300)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

ax.set_title('Spatial domain and boundary conditions')

airfoil.plot(ax)

x, y = [[-4.5, -4.5, -4.5], [-2.5, 0, 2.5]]
u, v = 2, 0
ax.quiver(x, y, u, v, scale=3, scale_units='xy', width=0.002, color=(0.8, 0.2, 0.2), label='$u_{in} = 15m/s$')

x, y = [[0, 5, 10, 0, 5, 10], [-5, -5, -5, 5, 5, 5]]
ax.scatter(x, y, color=(0.2, 0.8, 0.2), label='$v_{b} = 0m/s$')

x, y = [[15, 15, 15], [-2.5, 0, 2.5]]
ax.scatter(x, y, color=(0.2, 0.2, 0.8), label='$p_{out} = 1atm$')

domain = patches.Rectangle((-5, -5), 20, 10, linewidth=1, edgecolor=(0.5, 0.5, 0.5), facecolor='none', label='Domain')
ax.add_patch(domain)

ax.grid(False)
ax.set_aspect('equal')
ax.legend()
ax.set_xlabel('$x (m)$')
ax.set_ylabel('$y (m)$')
plt.savefig('naca_2412_2.png', dpi=300)
plt.show()
```
