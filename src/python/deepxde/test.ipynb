```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
loss = pd.read_csv('loss.dat', delimiter=' ', header=None)
steps = loss[0]
```

```python
fig, axs = plt.subplots(6, 2, figsize=(20, 25))

linewidth = 0.5

start_step = 100

for i in range(1, (int)((len(loss.columns)-1)/2)):
    axs[int((i-1)/2), (i-1) % 2].plot(steps[start_step:], loss[i][start_step:], linewidth=linewidth)
    axs[int((i-1)/2), (i-1) % 2].set_title('Loss ' + str(i))
```
