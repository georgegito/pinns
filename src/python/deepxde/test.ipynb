```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
loss = pd.read_csv('./out/loss.dat', delimiter=' ', header=None)
steps = loss[0]
```

```python
fig, axs = plt.subplots(6, 2, figsize=(20, 25))

linewidth = 0.5

for i in range(1, (int)((len(loss.columns)-1)/2)):
    axs[int((i-1)/2), (i-1) % 2].plot(steps, loss[i], linewidth=linewidth)
    axs[int((i-1)/2), (i-1) % 2].set_title('Loss ' + str(i))
```

```python
fig, axs = plt.subplots(6, 2, figsize=(20, 25))

linewidth = 0.5

for i in range(1, (int)((len(loss.columns)-1)/2)):
    axs[int((i-1)/2), (i-1) % 2].plot(steps, loss[(int)(i+len(loss.columns-1)/2+1)], linewidth=linewidth)
    axs[int((i-1)/2), (i-1) % 2].set_title('Loss ' + str(i))
```
