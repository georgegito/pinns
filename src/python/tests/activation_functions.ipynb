```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def swish(x):
    return x * sigmoid(x)

x = np.linspace(-10, 10, 100)

y_relu = relu(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_swish = swish(x)

# Plotting
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(x, y_relu, label="ReLU", linewidth=3)
plt.title("ReLU Activation Function")
plt.grid(False)

plt.subplot(2, 2, 2)
plt.plot(x, y_sigmoid, label="Sigmoid", linewidth=3)
plt.title("Sigmoid Activation Function")
plt.grid(False)

plt.subplot(2, 2, 3)
plt.plot(x, y_tanh, label="Tanh", linewidth=3)
plt.title("Tanh Activation Function")
plt.grid(False)

plt.subplot(2, 2, 4)
plt.plot(x, y_swish, label="Swish", linewidth=3)
plt.title("Swish Activation Function")
plt.grid(False)

plt.tight_layout()
plt.show()
```
